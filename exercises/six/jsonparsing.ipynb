```python
#sample data from: https://www.kaggle.com/gunnvant/game-of-thrones-srt

import json
import pandas as pd

with open("season1.json", "r") as read_file:
    data = json.load(read_file)

data_frame = pd.DataFrame(data)
data_frame
```

```python
reordered = data_frame.T
reordered
```
